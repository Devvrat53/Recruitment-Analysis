```python
df = pd.DataFrame(columns= ['username', 'description', 'location', 'following', 
                            'followers', 'text', 'likescount', 'retweetcount', 'date-time', 'hashtags'])
words = '#hiring'
date_since = '2020-05-07' #yyyy-mm-dd
numtweet = 2000

csvFile = open('scraped_tweets.csv', 'a')
csvWriter = csv.writer(csvFile)
```

```python
hashtext = list()
for tweet in tweepy.Cursor(api.search, q= words, lang= "en", since= date_since, tweet_mode='extended').items(numtweet):
    hashtags = tweet.entities['hashtags']
    for j in range(0, len(hashtags)):
        hashtext.append(hashtags[j]['text'])
    csvWriter.writerow([tweet.user.screen_name.encode('utf-8'), tweet.user.description.encode('utf-8'), 
                        tweet.user.location.encode('utf-8'), tweet.user.friends_count, tweet.user.followers_count, 
                        tweet.full_text.encode('utf-8'), tweet.favorite_count, tweet.retweet_count, tweet.created_at, 
                        hashtext])
```
